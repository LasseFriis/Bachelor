```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import netCDF4 as nc

import numpy as np
from matplotlib.animation import FuncAnimation
from IPython import display
import matplotlib.pyplot as plt
```

```python
fn = '../Data/iceh.2011-01.nc'
ds = nc.Dataset(fn)
ds['vicen'][0,0,292,292]
```

```python
fig, axes = plt.subplots(nrows=2, ncols=3)
for ax in axes.flat:
    im = ax.imshow(np.random.random((10,10)), vmin=0, vmax=1)

fig.subplots_adjust(right=0.8)
cbar_ax = fig.add_axes([0.85, 0.15, 0.05, 0.7])
fig.colorbar(im, cax=cbar_ax)

plt.show()
```

```python
plt.figure(figsize=(14,8))

plt.subplot(2,3,1)
plt.contourf(ds['vicen'][0,0], vmin=0, vmax=1)
plt.colorbar(extend='both')

plt.subplot(2,3,2)
plt.contourf(ds['vicen'][0,1], vmin=0, vmax=1)

plt.subplot(2,3,3)
plt.contourf(ds['vicen'][0,2], vmin=0, vmax=1)

plt.subplot(2,3,4)
plt.contourf(ds['vicen'][0,3], vmin=0, vmax=1)

plt.subplot(2,3,5)
plt.contourf(ds['vicen'][0,4], vmin=0, vmax=1)


plt.colorbar(extend='both')
plt.clim(0,1)
plt.show()
```

```python
plt.figure(figsize=(10,10))
fig, axes = plt.subplots(nrows=2, ncols=2)
i = 0
for ax in axes.flat:
    im = ax.imshow(ds['vicen'][0,i], vmin=0, vmax=1)
    i += 1
    
fig.subplots_adjust(right=0.8)
cbar_ax = fig.add_axes([0.85, 0.15, 0.05, 0.7])
fig.colorbar(im, cax=cbar_ax)

plt.show()
```
